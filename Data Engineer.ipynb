```python
import geopandas as gpd

# Load the Batam shapefile
batam_shapefile = 'shp batam/shpbatam.shp'
batam_geo = gpd.read_file(batam_shapefile)

# Check the CRS
print("Shapefile CRS:", batam_geo.crs)

# Assuming hotel_geo is your hotel GeoDataFrame
print("Hotel Data CRS:", hotel_geo.crs)

# If the CRS differs, change the hotel data CRS to match the shapefile
if hotel_geo.crs != batam_geo.crs:
    hotel_geo = hotel_geo.to_crs(batam_geo.crs)
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# Assuming your DataFrame is named 'df' and it has columns 'Latitude' and 'Longitude'
# Load your DataFrame (Replace with your data loading code)
# df = pd.read_csv('path_to_your_data.csv')

# Convert latitude and longitude to float
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')

# Create Point geometries
geometry = [Point(xy) for xy in zip(df['Longitude'], df['Latitude'])]
hotel_geo = gpd.GeoDataFrame(df, geometry=geometry)

# Set the CRS for hotel data to 'EPSG:4326' (standard for GPS coordinates)
hotel_geo.set_crs("EPSG:4326", inplace=True)

# Load the Batam shapefile
# Replace 'path_to_your_shapefile.shp' with the path to your shapefile
batam_shapefile = 'shp batam/shpbatam.shp'
batam_geo = gpd.read_file(batam_shapefile)

# Buffer the boundaries of the shapefile slightly
# Adjust the buffer distance (0.01 is an example)
buffered_batam_geo = batam_geo.buffer(0.01)

# Create a new GeoDataFrame with the buffered geometry
buffered_batam_geo_df = gpd.GeoDataFrame(geometry=buffered_batam_geo)

# Perform the spatial join to filter hotels within the buffered Batam boundaries
hotels_in_batam = gpd.sjoin(hotel_geo, buffered_batam_geo_df, op='within')

# hotels_in_batam now contains hotels within Batam, considering the buffer
hotels_in_batam
```

```python
hotels_in_batam.to_excel("filtered_data_scrapp_fix.xlsx")  
```

```python
# Remove duplicates based on the unique identifier
hotels_in_batam_cleaned = hotels_in_batam.drop_duplicates(subset=['Link'])

# Check the number of rows after cleaning
print(f"Number of rows after removing duplicates: {len(hotels_in_batam_cleaned)}")
```

```python
tipe = hotels_in_batam_cleaned['Tipe Konstruksi'].str.contains('CV|PT|kontraktor|kontruksi|konstruksi|Construction', case=False, na=False)

# Create a boolean series where 'kegiatan_utama' contains the desired strings
name = hotels_in_batam_cleaned['Name'].str.contains('kontraktor|kontruksi|konstruksi', case=False, na=False)


# Combine both conditions with an OR logic
combined_condition = tipe | name

# Filter the DataFrame using the combined condition
filtered_data_bps = hotels_in_batam_cleaned[combined_condition]

# Display the first few rows of the filtered DataFrame
filtered_data_bps.head()
```
