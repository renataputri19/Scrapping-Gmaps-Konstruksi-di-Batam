```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
import time
```

```python
# Path to the ChromeDriver
chrome_driver_path = 'chromedriver-win64/chromedriver.exe'

# Path to the Chrome binary
chrome_binary_path = 'chrome-win64/chrome.exe'

# Setting up Chrome options
chrome_options = Options()
chrome_options.binary_location = chrome_binary_path
```

```python
# Initializing the ChromeDriver
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
konstruksi_latitude, konstruksi_longitude = [], []
konstruksi_name = []
konstruksi_alamat, konstruksi_telfon, konstruksi_website = [], [], [] 
construction_tipe = []


for links in konstruksi_links_tes:
    
    alamat, telfon, website = "", "", ""  # Initialize with empty strings
    tipe = ""  # Initialize with default values
    
    try:
        driver.get(links)

        # Waiting for the konstruksi page to load
        time.sleep(5)

        current_url = driver.current_url

        # Extracting latitude and longitude from the URL using a regular expression
        coordinates = re.search(r'@(-?\d+\.\d+),(-?\d+\.\d+)', current_url)
        if coordinates:
            latitude = coordinates.group(1)
            longitude = coordinates.group(2)
#             print(f"Latitude: {latitude}, Longitude: {longitude}")
            konstruksi_latitude.append(latitude)
            konstruksi_longitude.append(longitude)

        try:
            # Find the element by class name
            # Wait for the elements to be present
            WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "DUwDvf")))
            konstruksi_name_element = driver.find_element(By.CLASS_NAME, 'DUwDvf')
            name = konstruksi_name_element.text
#             print(f"konstruksi Name: {name}")
            konstruksi_name.append(name)

        except Exception as e:
#             print(f"Error retrieving konstruksi name: {e}")
            konstruksi_name.append("")

        max_attempts = 3  # Define the maximum number of retries
        
        try:
            WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "CsEnBe")))
            details_elements = driver.find_elements(By.CLASS_NAME, 'CsEnBe')

            for attempt in range(max_attempts):             

                for element in details_elements:
                    WebDriverWait(driver, 10).until(lambda d: element.get_attribute("aria-label").strip() != "")
                    aria_label = element.get_attribute("aria-label") if element.get_attribute("aria-label") else ""

                    if "Alamat:" in aria_label:
                        WebDriverWait(driver, 10).until(lambda d: element.text.strip() != "")
                        alamat = element.text.strip()
                    elif "Telepon:" in aria_label:
                        WebDriverWait(driver, 10).until(lambda d: element.text.strip() != "")
                        telfon = element.text.strip()
                    elif "Situs Web:" in aria_label:
                        WebDriverWait(driver, 10).until(lambda d: element.text.strip() != "")
                        website = element.text.strip()

                # Check if at least one of the details is present
                if alamat or telfon or website:
                    # At least one detail is present, append to the lists
                    break  # Data found, exit retry loop
                else:
                    # If data is missing, wait and then retry
#                     print(f"Attempt {attempt + 1}: Missing details, retrying...")
                    time.sleep(2)  # Wait before retrying

                if attempt == max_attempts - 1:
                    # If all attempts fail, append whatever data is available
#                     print("All attempts failed to retrieve full details. Proceeding with available data.")
                    time.sleep(2)
                    
#         konstruksi_alamat.append(alamat)
#         konstruksi_telfon.append(telfon)
#         konstruksi_website.append(website)                


        except Exception as e:
#             print(f"All attempts failed: {e}")
            time.sleep(2)


        max_attempts = 3  # Define the maximum number of retries

        try:
            for attempt in range(max_attempts):
                try:

                    # Wait for the button to be present
                    button = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.CLASS_NAME, "DkEaL"))
                    )

                    # Extract and print the button text
                    tipe = button.text


                    driver.back()
                    time.sleep(3)  # Wait for the list to load
                    break  # Break the loop if successful

                except Exception as e: 
                    time.sleep(2)
#                     print(f"Attempt {attempt + 1} failed: {e}")
                      # Wait a bit before retrying

                    if attempt == max_attempts - 1:
                        raise  # Raise the last exception if all attempts fail
                    
#         konstruksi_harga.append(harga)
#         konstruksi_bintang.append(bintang)

        except Exception as e:
#             print(f"All attempts failed: {e}")
            time.sleep(2)
        
        # Go back to the list of konstruksis
        driver.back()
        time.sleep(3)  # Wait for the list to load
    except:
        print("error gan awal")
        
    konstruksi_alamat.append(alamat)
    konstruksi_telfon.append(telfon)
    konstruksi_website.append(website) 
    
    construction_tipe.append(tipe)
```
